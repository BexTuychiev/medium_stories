### Introduction

Christmas has been canceled for Python. After becoming the "Language of the Year" three times in four years, it looks like it is going to lose the award spectacularly this year. And the language that is snatching it away is C#.

But why then, the image shows that Python is at the top while C# is only in 5th place? Looks like Python is winning.

Well, to fully answer the question, we should understand the TIOBE index and how its ranking works.

### TIOBE index

TIOBE is a software quality company founded in 2000. It specializes in assessing the quality of software systems and processes more than a billion lines of code each day.

But that's not what we care about, is it? Among general programming population, TIOBE is mainly known for [its language index](https://www.tiobe.com/tiobe-index/):



Its website shows the growth of more than 150 languages going as far back as to 80s:


You will see that titan languages like C, Java and C++ have shared the glorious spot since 1987. And it gives the most pleasure to see that Python kicked them all cleanly out of the ballpark after being at the bottom of the barrel in the beginning. 

But apparently, being at the top with a significant margin isn't enough to become the "Language of the Year" every year. 

So, what does it take?

### How is the TIOBE index calculated?

Let's start with how the global rankings (the Ratings column of the below table) are calculated:


You will be shocked at how easy it is to determine. The TIOBE index just calculates the proportion of search engine queries containing the name of a language.

TIOBE only considers 25 search engines and here are the top 15:



To calculate Python's, the index counts the query "+Python programming" in each 25 engines, and divides by the total number of programming questions to get a percentage value. 

So, 12.74% rating means that Python questions account for about 13% of all programming-related queries in search engines. In other words, one in eight of every programming question on the Internet is about Python. Given that the index includes more than 150 languages, this is a significant feat for Python. 

To qualify as a language in the index, the main requirement is that the language must have its own Wikipedia entry clearly stating that it is a programming language. The names like Ruby on Rails, Excel, Android and AJAX aren't in the index because they fall short of this requirement.

### So, why is C# winning?

To win the award, TIOBE index looks at the percentage growth of the language since last year, not its actual place in the rating. The June index 2022 shows that the number of C# questions on the internet grew by about 2%:



Python grew by about 1% which is even lower than other languages like C++ and Visual Basic (this one hurts, by the way).

The year isn't over yet but it seems highly unlikely that Visual Basic is going to catch up with C# and Python is completely out of the running. 

C# will win it for the first time ever since its release in 2000. So, congratulations to all .NET programmers.
