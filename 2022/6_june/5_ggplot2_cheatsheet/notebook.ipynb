## Customizing visualizations with ggplot2

> Axis labels and scale customization (please add at least 3 to 5 functions that allow practitioners to customize the axis of their visualizations, such as changing axis scale, updating axis labels, changing ticks)

##### 1. Switching to logarithmic scale

```r
scatter = ggplot(diamonds, aes(price, carat))
scatter + geom_point() + scale_x_log10() # or scale_y_log10
```

##### 2. Reverse the direction of the axis

```r
scatter = ggplot(diamonds, aes(price, carat))
scatter + geom_point() + scale_x_reverse()
```

##### 3. Square root scale

```r
scatter = ggplot(diamonds, aes(price, carat))
scatter + geom_point() + scale_x_sqrt()
```

##### 4. Changing axis limits with clipping

```r
plot = ggplot(diamonds, aes(table, depth)) + geom_point()
plot + coord_cartesian(xlim=c(55, 70), ylim=c(55, 75))
```

##### 5. Changing axis limits with clipping

```r
plot = ggplot(diamonds, aes(table, depth)) + geom_point()
plot + xlim(55, 70) + ylim(55, 75)
```

##### 6. Changing labels on the plot

```r
plot = ggplot(diamonds, aes(table, depth)) + geom_point()

plot + labs(x='Diamond table', y='Diamond depth',
           title='New title',
           subtitle='Sample subtitle',
           caption='Image by author')
```

> Adding legend (please add at least 3 to 5 functions that allow practitioners to customize the legend, such as setting position, updating fonts and size, etc..)

```r
scatter = ggplot(diamonds, aes(price, carat, size=table, color=depth))
scatter = scatter + geom_point()
```

```r
scatter + labs(color="Diamond depth", size="Diamond table")
```

```r
scatter + theme(legend.position="none")
```

##### 4. Change legend position outside of the plot

```r
# You can also pass top, right, left
scatter + theme(legend.position = "bottom")
```

##### 5. Placing the legend into the plot area:

```r
scatter + theme(legend.position=c(.1, .7))
```

## Facetting

> Please add a 2-3 sentence description of what is Facetting

Faceting is a data visualization technique to explore the dataset more granularly using categorical columns. For example, if a column has three categories a, b, c, faceting by that column would create three figures for each group and the user would be able to fill them by plotting other variables.

> Please provide a simple example of facet_wrap in action

##### 1. Facet the figures into a rectangular layout of two columns

```r
scatter = ggplot(diamonds, aes(price, carat)) + geom_point()
scatter + facet_wrap(vars(cut), ncol=2)
```

##### 2. Facet the figures into a rectangular layout of two rows

```r
scatter = ggplot(diamonds, aes(price, carat)) + geom_point()
scatter + facet_wrap(vars(cut), nrow=2)
```

##### 3. Facet into a rectangular layout but give axes free ranges (variable plot dimensions):

```r
scatter = ggplot(diamonds, aes(price, carat)) + geom_point()
scatter + facet_wrap(vars(cut), scales="free")
```

##### 4. Faceting by both columns and rows with `facet_grid`

```r
scatter + facet_grid(rows=vars(clarity), cols=vars(cut),
                    space='free', scales='free')
```
