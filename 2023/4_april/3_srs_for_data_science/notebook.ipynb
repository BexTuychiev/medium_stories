### Brain dump

- I believe that self-taught data scientists are the worst type of learners
- There are no pressures like university exams or regular tests so that you keep your spot in some institution
- Information about ds is freely available everywhere so its perceived value is very low
- People mistake watching youtube tutorials or taking udemy courses for active learning
- When actually, there is very little learning going on
- The way we, self-taught data scientists study, is very haphazard
- We always leave memory up to chance. 
- We learn something once and hope for the best that it stays there
- Or think that we will come back to those concepts once again later
- most of what we learn are forgotten for lack of pressure to do well and ace those subjects
- So, they keep watching the same tutorials and taking the courses again and again
- Yes, there are people who practice with coding
- But even those spend a lot of their time googling concepts they already learned and googled dozens of times before because they forgot
- And they brag to others how they are so "good at Googling" stuff while they are actually subtly signaling we have no systems whatsoever to remember all the overwhelming amount of information we learn in data science
- I believe it is not something to be proud of but rather, it is something we should fix
- Spaced repetition totally changes that
- Once you put what you learned into a reliable repetition system, you are consciously making a choice to remember that piece of info for the rest of your life or as long as you need it.
- Without an effective way to learn and retain new information, it is very hard to become a data scientist
- Learning new things is easy most of the time. 
- Retaining those things in memory is totally different subject

- The single most effective way to study is spaced repetition
- In sr, you review information in increasingly larger intervals
- With sr, you can memorize new information in much less time that it would take otherwise
- The way most of us study is in long sessions where we try to cram everything into our minds 
- Pull all-nighters
- Memories in the brain becomes strong by deliberate, active recalling and repetition, which is what sr is all about.
- Re-exposes you to information at optimal intervals, just when you are about to forget it
- It increases the amount of active practice
- 
- It is based on the Ebbinghaus forgetting curve, which shows how much new information leaks from your brain as time passes
- Spaced repetition is regularly associated with flashcards
- Flashcards can be used to store little bits of information to be memorized at a time
- This makes synthesizing new information easier
- But that's the biggest problem in spaced repetition
- It becomes tedious to create 100s of flashcards and organize them
- If spaced repetition creates worse problems like organizing 100s of flashcards and waste your time, it wouldn't be worth your attention
- But thankfully, there are tools
- First of all, there is Anki, an open source spaced repetition flashcards software. 
- It has many pre-built decks for subjects like geography, math, medicine, statistics or for most subjects that appear in university exams
- SRS tools allow you to create your own cards and keeps track of how well you memorize them
- This frees you from the manual work of keeping note of which flashcards you have to review every day
- Personally, I married AnkiPro.net for life. It costs 5$ a month and has less features than Anki, but it has a modern minimal interface and has pre-built decks that I want
- But data science has overwhelming amount information to learn:
    - math
    - stats
    - programming concepts
    - thousands of parameters, methods and functions of python frameworks
    - interview preparation and so on
- You can't possibly create flashcards for all these topics
- Thankfully, we are all in luck to be living in the golden age of AI (we are, aren't we? I still can't believe it)
- There are now AI tools to automatically create flashcards from any information you provide
- One of the tools I found was monic.ai 
- It is either short for Monica AI or mnemonic AI, depending on your psychological orientation. 
- You just upload a pdf or image of the text you just learned and it automatically creates the flashcards for you.
