# Forget PIP, Conda, requirements.txt! Use Poetry Instead And Thank MeLater
## Pain-free dependency management is finally here


### Introduction

- Data scientists always rely on external packages
- No built-in alternatives
- And when there are dependencies, there is a dependency hell
- The case when library A requires a specific Python version when another needs a better while also requiring that same version of library that needs older python version
- Welcome to hell!
- Tools like PIP, Conda or our laughable requirements.txt can't solve this problem. 
- Actually, dependency nightmares exist largely _because of them_
- So, to end their suffering, Python open-source developers wrote charming Poetry. It is a project and dependency management framework.
- This article will list the problems it solves for data scientists.

### Installation

Linux, macOS, Windows (WSL):

```
curl -sSL https://install.python-poetry.org | python3 -
```
