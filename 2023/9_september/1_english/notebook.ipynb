How well you express yourself in English (not another language) is one of the biggest precursors to your success in machine learning. It is not math, statistics or even programming. It is fluent, unflinching, powerful command of English.

With fluent English, you can learn the vast amount information required for an ML job, network like a son of a gun with recruiters and CEOs, not stutter your way into failure during interviews, gain the respect of your colleagues in projects, negotiate your salary and ask for promotions without embarrassing yourself, advance your skill set and career by reading SOTA research papers and participating in conferences.

### Learning and growing your knowledge base

Despite the admirable efforts of online courses and blogs to simplify machine learning education, it still remains quite academic. 

While most online pages and videos on any other topic has a 6-7th grade average, any respectable machine learning resource has at least 10th-grade reading level. With the mathematics factor added, machine learning quickly becomes a top-tier profession in terms of mastery (that's why we get paid big bucks). 

But most ML training programs don't represent this harsh reality. To increase sales and reduce churn, they keep dumbing down their content to give you the illusion that all you need in ML is Python and a few other libraries to secure a paid position. 

When aspiring engineers, who are given a confidence boost by their cheap Udemy certificate, try to tackle real-world projects, it is like they are transported into a whole other world. Their English comprehension, which was more than enough for Udemy, becomes laughably insufficient.

So, they start littering QA sites or other people's DMs with easily _googleable_ or _promptable_ questions (exhibit A in the introduction).  

The information required for an ML job is like a vast and turbulent sea. To swim across it, you need a good life jacket. That life jacket is how well you know English.

### Building your portfolio

When beginners start thinking about landing their first role, they start building the highly-recommended "stellar portfolio".

But they go at it the wrong way and fill their GitHub profiles with complicated deep learning models. They think "The more complex, the better".

But it is not about the code (not 100%). Today, anyone can write high-quality, production-level code with decent prompt engineering techniques. In this age of LLMs, writing code lost its weight when building portfolios that land interviews. 

It is often about how a person can come up with simple, original solutions to existing ML problems. And more importantly, how well they can present their unique solution.

If you can't explain your thought process, methodology - your project effectively and concisely in a GitHub README, you might've as well copy-pasted the entire thing.

Writing READMEs is easier said than done. If it wasn't, every single person on GitHub would have repos with thousands of stars. 

For the longest time, I thought people were just too lazy to create a decent README to summarize their work. Turns out, it is actually more about lacking the writing skills to put together a cohesive paragraph. 

Jeff Bezos said, "There is no way to write a six-page narratively structured memo and not have clear thinking". A machine learning position at Amazon is a dream for most beginners in our field, and look at what's asked from its employees in the workplace. 

### Surviving interviews

In the fiercely competitive ML marketplace, landing interviews is already a huge step for beginners. Interviews are all about you __talking__ non-stop to prove your worth for the applied position. 

This is where English is the last thing you should be thinking about. Your spoken English should already be awesome before you shake hands with the interviewers. 

During the actual interview, you have dozens of things to worry about - the dozen things listed here. If you are thinking about how to communicate the ideas in your head, instead of delivering them on autopilot, you are gonna put too much _ums_, _ehhs_, talk without any intonation, make mistakes, blurt out incomprehensible statements and drive the interviews into half-asleep.

Every fluent sentence you say counts towards beating the hundreds of applicants, waiting for you to fail like sharks.
